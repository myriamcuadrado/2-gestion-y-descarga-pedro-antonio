```python
# Altair_Fig3 --> hacer interactiva al presentar el ćodigo. 

source = df2
domain = ['CCND1', 'CCND2', 'CCND3']
range = ['#17becf', '#fb8072', '#b3cde3']

alt.Chart(source).transform_filter(
    alt.datum.symbol != 'GOOG'
).mark_area().encode(
    x='SAMPLE:T',
    y='EXPRESSION:Q',
    color= alt.Color('Cyclin:N', scale=alt.Scale(domain=domain, range=range )),
    row=alt.Row('Cyclin:N', sort=['CCND1', 'CCND2', 'CCND3'])
).properties(height=100, width=500)
```

```python
# Altair_Fig4 --> hacer interactiva al presentar el ćodigo. 

alt.Chart(source).mark_circle(
    opacity=0.5,
    stroke='gray',
    strokeWidth=0.3
).encode(
    alt.X('EXPRESSION:O', axis=alt.Axis(labelAngle=0)),
    alt.Y('Cyclin:N'),
    alt.Size('SAMPLE:Q',
        scale=alt.Scale(range=[0, 200]),
        legend=alt.Legend(title='EXPRESSION')
    ),
    alt.Color('Cyclin:N', scale=alt.Scale(domain=domain, range=range ))
).properties(
    width=600,
    height=250
)
```
